- Exercici 4
Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

```python
import numpy as np

x = np.array([[54, 23], [23, -90], [25, -43]])
print(x.shape)
print('x pot ser broadcasted amb 10, ja que la seva mida és = 1: ', x + 10)

y = np.array([[1, 2], [3, 4], [5, 6]])
print(y.shape)

print('x i y es poden transmetre o fer broadcast: ', x * y)

z = np.array([50, 100])
print(z.shape)

print('x i z es poden transmetre o fer broadcast:', x - z)

a = np.array([[500], [400]])
a_nou = a.reshape(2,)
print(a_nou.shape)
print('a_nou i y es poden transmetre gràcies al reshape, si no no podrien: ', a_nou + y)

b = np.array([54, 23, 23, -90, 25, -43])
b_nou = b.reshape(3, 2)
print(b_nou.shape)
print('b_nou i x es poden transmetre gràcies al reshape, si no no podrien:', b_nou + x)

c = a[:, np.newaxis] * b
print('També es pot fer broadcast en un altre eix:', c)

print('A continuació sortirá un error degut a que les dos matrius no comparteixen característiques que els permeti fer broadcast:')
print(x*b)
      
```

- Exercici 5
Utilitza la Indexació per extreure els valors d'una columna i una fila de la matriu. I suma els seus valors.

```python
import numpy as np
x = np.arange(9)
matr_x = x.reshape(3, 3)
print(matr_x)

print('Segona columna de la matriu:', matr_x[:, 1])
print('Tercera fila de la matriu:', matr_x[2, :])
print('Suma de la fila y la columna:', matr_x[:, 1] + matr_x[2, :])
print('Suma total de la fila y la columna:', np.sum(matr_x[:, 1] + matr_x[2, :]))
```

- Exercici 6
Mask la matriu anterior, realitzeu un càlcul booleà vectoritzat, agafant cada element i comprovant si es divideix uniformement per quatre.

Això retorna una matriu de mask de la mateixa forma amb els resultats elementals del càlcul.

```python
import numpy.ma as ma
print('Matriu original:\n', matr_x)

ma_x = ma.masked_where(matr_x % 4 == 0, matr_x).mask
print('Mask:\n', ma_x)
```
